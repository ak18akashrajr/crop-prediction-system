```python
import pandas as pd
df=pd.read_csv("weatherHistory.csv")
df
```

```python
my_list = list(df)
my_list
```

```python
check_nan = df['Precip Type'].isnull().sum()
check_nan
```

```python
df.rename(columns = {'Formatted Date':'date', 'Summary':'summary',
                              'Precip Type':'type'}, inplace = True)
df
```

```python
df=df.drop(['date', 'summary','Wind Bearing (degrees)','Loud Cover','Daily Summary'], axis=1)
df
```

```python
df.rename(columns = {'Temperature (C)':'temp', 'Apparent Temperature (C)':'app_temp',
                              'Wind Speed (km/h)':'w_speed','Humidity':'humidity','Visibility (km)':'visibility',
                              'Pressure (millibars)':'pressure'}, inplace = True)
df
```

```python
check_nan = df['type'].isnull().sum()
check_nan
```

```python
df["type"].fillna("sunny", inplace = True)
```

```python
check_nan = df['type'].isnull().sum()
check_nan
```

```python
weather_mapping = {'rain': 1, 'sunny': 0}
df['type'] = df['type'].map(weather_mapping)
```

```python
df
```

```python
X=df[['temp','app_temp','humidity','w_speed','visibility','pressure']]
X
```

```python
Y=df[['type']]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python

from sklearn import tree
from sklearn.tree import plot_tree
model=tree.DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
pruning=tree.DecisionTreeClassifier(max_depth = 2)
pruning.fit(X_train,Y_train)
plot_tree(pruning,filled=True)
a=pruning.score(X_test,Y_test)
a
```

```python
from sklearn.ensemble import RandomForestRegressor
 
# create regressor object
regressor = RandomForestRegressor(n_estimators=100,
                                  random_state=0)
 
# fit the regressor with x and y data
regressor.fit(X_train, Y_train)
```

```python
b=regressor.score(X_test,Y_test)
b
```

```python
import joblib 
# Save the model as a pickle in a file 
joblib.dump(regressor, 'rf.pkl') 
```

```python
import matplotlib.pyplot as plt  
# data to display on plots  
x = ["Decision Tree","Random Forest"]  
y = [a,b]  
  
# This will plot a simple bar chart 
plt.bar(x, y) 
  
# Title to the plot 
plt.title("Bar Chart") 
  
# Adding the legends 
plt.legend(["bar"]) 
plt.show()
```
