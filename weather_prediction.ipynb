```python
import pandas as pd
df=pd.read_csv("weatherHistory.csv")
df
```

```python
my_list = list(df)
my_list
```

```python
check_nan = df['Precip Type'].isnull().sum()
check_nan
```

```python
df.rename(columns = {'Formatted Date':'date', 'Summary':'summary',
                              'Precip Type':'type'}, inplace = True)
df
```

```python
df=df.drop(['date', 'summary','Wind Bearing (degrees)','Loud Cover','Daily Summary','type'], axis=1)
df
```

```python
df.rename(columns = {'Temperature (C)':'temp', 'Apparent Temperature (C)':'app_temp',
                              'Wind Speed (km/h)':'w_speed','Humidity':'humidity','Visibility (km)':'visibility',
                              'Pressure (millibars)':'pressure'}, inplace = True)
df
```

```python
check_nan = df['visibility'].isnull().sum()
check_nan
```

```python
X=df[['temp','app_temp','w_speed','visibility','pressure']]
X
```

```python
Y=df[['humidity']]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.1,random_state=42)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X_train,Y_train)
b=model.score(X_test,Y_test)
b
```

```python
model.predict(X_test)
```

```python
from sklearn import linear_model
ridge = linear_model.Ridge(alpha=10)
ridge.fit(X_train,Y_train)
training_ridge = ridge.score(X_test,Y_test)
print("training accuracy with ridge ",training_ridge )
```

```python
import matplotlib.pyplot as plt  
# data to display on plots  
x = ["Ridge Regression","Linear Regression"]  
y = [training_ridge,b]  
  
# This will plot a simple bar chart 
plt.bar(x, y) 
  
# Title to the plot 
plt.title("Bar Chart") 
  
# Adding the legends 
plt.legend(["bar"]) 
plt.show()
```
